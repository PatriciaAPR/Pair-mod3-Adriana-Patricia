```python
x= '652,230' #Ejemplo César (fila 1 de columna land_area)
```

```python
def cambiar (dato):                 # Definimos la función
    try:
        dato = dato.replace(',', '')
        return float(dato)
    except:
        return np.nan       # Para los nulos
```

```python
cambiar(x) # Probamos con el ejemplo
```

```python
df['land_area'].apply(cambiar) # Aplicamos a toda la columna
```

```python
columnas = ['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitude', 'longitude', 'density', 'cpi'] # Lista columnas

for columna in columnas:                # Aplicamos la función a todas las columnas mencionadas con el bucle for
    df[columna] = df[columna].apply(cambiar)


df.info()

```
